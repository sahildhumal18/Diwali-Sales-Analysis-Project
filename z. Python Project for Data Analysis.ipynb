```python
# drop the blank column from the dataframe - 
# use inplace = true opreation. It will modify the dataframe directly without creating a new dataframe.
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
# seperate the categorical data and continous data from the dataframe - 
def categorical_continuous_data_seperation(df):
    categorical_data = []
    continuous_data = []
    
    for i in df.columns:
        if (df[i].dtype == 'object'):
            categorical_data.append(i)
            
        else:
            continuous_data.append(i)
            
    return categorical_data, continuous_data
```

```python
categorical_data, continuous_data = categorical_continuous_data_seperation(df)
```

```python
# check the null values in the dataframe - 
# .sum() returns the total of null values
pd.isnull(df).sum()
```

```python
# replace null values present in the dataframe - 
df['Amount'].fillna(df['Amount'].mean(), inplace=True)
```

```python
# change the datatype of the amount from float to integer - 
df['Amount'] = df['Amount'].astype('int')
```

```python
a = sns.countplot(x='Gender', data=df)
plt.title("Total count of males and females in dataframe")
# to label the bars with count use the following method - 
for bars in a.containers:
    a.bar_label(bars)
```

```python
# analysis of total amount spend by each gender - 
sale = df.groupby(['Gender'], as_index=False)['Amount'].sum()
sale = sale.sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sale)
```
